Import

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Read Csv

```python
df = pd.read_csv('Files/who_suicide_statistics.csv')

df
```

Remover Colunas desnecessarias e deletar valores nulos

```python
df = df.drop(columns=['population', 'suicides_no'])
df
```

Convertendo Tipo da coluna year

```python
df['year'] = pd.to_numeric(df['year'])
df
```

Agrupando dados por ano

```python
data = df.groupby(['year', 'sex']).size().unstack()
data['total'] = data['male'] + data['female']
data
```

Exibindo valores com Matplotlib

```python
fig, ax = plt.subplots()

x = data.index
for column in data.columns:
    ax.plot(x, data[column], label = column, marker = 'o')
    plt.legend()
```
